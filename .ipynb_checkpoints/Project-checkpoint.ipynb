```python
# Importing Needed Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error , mean_squared_log_error

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('world.csv', decimal=',')
print('Number Of Missing Data:')
print(df.isnull().sum())
```

### Data Preparation

#### Filling in Missing Values
To simplify the process of filling missing data, I propose to use the median values of the corresponding region for a given country. This approach is based on the fact that countries that share geological proximity tend to exhibit similarities in various aspects. In this case, I will consider the median values of the region for three variables: 'GDP ($ per capita)', 'Literacy (%)', and 'Agriculture'.

For the 'climate' feature, we will use the mode instead of the median since 'climate' is a categorical variable. The mode represents the most frequent category in the dataset, and it is a suitable measure of central tendency for categorical variables. Therefore, we will use the mode of the region for a given country to fill in any missing data for the 'climate' feature.

```python
df.groupby('Region')[['GDP ($ per capita)','Literacy (%)','Agriculture']].median()
```

```python
for col in df.columns.values:
    if df[col].isnull().sum() == 0:
        continue
    if col == 'Climate':
        guess_values = df.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
    else:
        guess_values = df.groupby('Region')[col].median()
    for region in df['Region'].unique():
        df[col].loc[(df[col].isnull())&(df['Region']==region)] = guess_values[region]
```

## Modeling

### Training and Testing
To prepare the categorical features 'Region' and 'Climate' for use in a machine learning model, we will apply label encoding. This technique assigns a unique numerical label to each category of a categorical variable, thereby converting it into a numerical feature that can be used in machine learning models.

We will use all the features provided in the dataset without further feature engineering.

```python
LE = LabelEncoder()
df['Region_label'] = LE.fit_transform(df['Region'])
df['Climate_label'] = LE.fit_transform(df['Climate'])
df.head()
```

```python
train, test = train_test_split(df, test_size=0.3, shuffle=True)
training_features = ['Population', 'Area (sq. mi.)', 'Pop. Density (per sq. mi.)', 'Coastline (coast/area ratio)',
                     'Net migration', 'Infant mortality (per 1000 births)', 'Literacy (%)', 'Phones (per 1000)',
                     'Arable (%)', 'Crops (%)', 'Other (%)', 'Birthrate', 'Deathrate', 'Agriculture', 'Industry', 
                     'Service', 'Region_label', 'Climate_label']
target = 'GDP ($ per capita)'
train_X = train[training_features]
train_Y = train[target]
test_X = test[training_features]
test_Y = test[target]
```

To begin, Let's use a linear regression model to make predictions based on the features in the dataset. We will evaluate the performance of the model using two metrics: root mean squared error (RMSE) and mean squared log error (MSLE). RMSE measures the average difference between the predicted and actual values, while MSLE measures the average squared difference between the logarithm of the predicted and actual values. By considering both metrics, we can gain a more comprehensive understanding of the performance of the model.

```python
model = LinearRegression()
model.fit(train_X, train_Y)
train_pred_Y = model.predict(train_X)
test_pred_Y = model.predict(test_X)
train_pred_Y = pd.Series(train_pred_Y.clip(0, train_pred_Y.max()), index=train_Y.index)
test_pred_Y = pd.Series(test_pred_Y.clip(0, test_pred_Y.max()), index=test_Y.index)

rmse_train = np.sqrt(mean_squared_error(train_pred_Y, train_Y))
msle_train = mean_squared_log_error(train_pred_Y, train_Y)
rmse_test = np.sqrt(mean_squared_error(test_pred_Y, test_Y))
msle_test = mean_squared_log_error(test_pred_Y, test_Y)

print('rmse_train:',rmse_train,'msle_train:',msle_train)
print('rmse_test:',rmse_test,'msle_test:',msle_test)
```

Since we know that the relationship between the target variable and the features may not be purely linear, it is worth exploring the use of non-linear models. One such model is the random forest algorithm, which is an ensemble learning method that combines multiple decision trees to make predictions. Random forest models are capable of capturing non-linear relationships between the features and the target variable, and they often outperform linear models when dealing with complex datasets.

```python
model = RandomForestRegressor(n_estimators = 50,
                              max_depth = 6,
                              min_weight_fraction_leaf = 0.05,
                              max_features = 0.8,
                              random_state = 42)
model.fit(train_X, train_Y)
train_pred_Y = model.predict(train_X)
test_pred_Y = model.predict(test_X)
train_pred_Y = pd.Series(train_pred_Y.clip(0, train_pred_Y.max()), index=train_Y.index)
test_pred_Y = pd.Series(test_pred_Y.clip(0, test_pred_Y.max()), index=test_Y.index)

rmse_train = np.sqrt(mean_squared_error(train_pred_Y, train_Y))
msle_train = mean_squared_log_error(train_pred_Y, train_Y)
rmse_test = np.sqrt(mean_squared_error(test_pred_Y, test_Y))
msle_test = mean_squared_log_error(test_pred_Y, test_Y)

print('rmse_train:',rmse_train,'msle_train:',msle_train)
print('rmse_test:',rmse_test,'msle_test:',msle_test)
```

### Visualization of Results

To assess the performance of the model, we can create a scatter plot of the predicted values against the actual values. A well-performing model will result in data points that cluster around the line y=x, indicating that the predicted values are close to the actual values.

```python
plt.figure(figsize=(18,12))

train_test_Y = train_Y.append(test_Y)
train_test_pred_Y = train_pred_Y.append(test_pred_Y)

data_shuffled = df.loc[train_test_Y.index]
label = data_shuffled['Country']

colors = {'ASIA (EX. NEAR EAST)               ':'red',
          'EASTERN EUROPE                     ':'orange',
          'NORTHERN AFRICA                    ':'gold',
          'OCEANIA                            ':'green',
          'WESTERN EUROPE                     ':'blue',
          'SUB-SAHARAN AFRICA                 ':'purple',
          'LATIN AMER. & CARIB                ':'olive',
          'C.W. OF IND. STATES                ':'cyan',
          'NEAR EAST                          ':'hotpink',
          'NORTHERN AMERICA                   ':'lightseagreen',
          'BALTICS                            ':'rosybrown'}

for region, color in colors.items():
    X = train_test_Y.loc[data_shuffled['Region']==region]
    Y = train_test_pred_Y.loc[data_shuffled['Region']==region]
    ax = sns.regplot(x=X, y=Y, marker='.', fit_reg=False, color=color, scatter_kws={'s':200, 'linewidths':0}, label=region) 
plt.legend(loc=4,prop={'size': 12})  

ax.set_xlabel('GDP ($ per capita) ground truth',labelpad=40)
ax.set_ylabel('GDP ($ per capita) predicted',labelpad=40)
ax.xaxis.label.set_fontsize(24)
ax.yaxis.label.set_fontsize(24)
ax.tick_params(labelsize=12)

x = np.linspace(-1000,50000,100) # 100 linearly spaced numbers
y = x
plt.plot(x,y,c='gray')

plt.xlim(-1000,60000)
plt.ylim(-1000,40000)

for i in range(0,train_test_Y.shape[0]):
    if((data_shuffled['Area (sq. mi.)'].iloc[i]>8e5) |
       (data_shuffled['Population'].iloc[i]>1e8) |
       (data_shuffled['GDP ($ per capita)'].iloc[i]>10000)):
        plt.text(train_test_Y.iloc[i]+200, train_test_pred_Y.iloc[i]-200, label.iloc[i], size='small')
```

Based on the scatter plot, we can see that the Random Forest Model is providing reasonable predictions, as the data points are tightly clustered around the y=x line.

### Total GDP

In addition to examining individual features and model performance, it is also useful to analyze the total GDPs of the countries. We can calculate the total GDP by multiplying the per capita GDP ('GDP ($ per capita)') by the population of the country.

After calculating the total GDPs, we can identify the top 10 countries with the highest total GDPs. These countries are likely to have a significant impact on the global economy, as they make up a large proportion of the world's total GDP. In fact, the total GDPs of the top 10 countries may account for as much as two-thirds of the global GDP.

```python
df['Total_GDP ($)'] = df['GDP ($ per capita)'] * df['Population']
#plt.figure(figsize=(16,6))
top_gdp_countries = df.sort_values('Total_GDP ($)', ascending=False).head(10)
other = pd.DataFrame({'Country':['Other'], 'Total_GDP ($)':[df['Total_GDP ($)'].sum() - top_gdp_countries['Total_GDP ($)'].sum()]})
gdps = pd.concat([top_gdp_countries[['Country','Total_GDP ($)']],other], ignore_index=True)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,7), gridspec_kw = {'width_ratios':[2,1]})
sns.barplot(x='Country',y='Total_GDP ($)', data=gdps, ax=axes[0], palette='Set3')
axes[0].set_xlabel('Country', labelpad=30, fontsize=16)
axes[0].set_ylabel('Total_GDP', labelpad=30, fontsize=16)

colors = sns.color_palette("Set3", gdps.shape[0]).as_hex()
axes[1].pie(gdps['Total_GDP ($)'], labels=gdps['Country'], colors=colors, autopct='%1.1f%%',shadow=True)
axes[1].axis('equal')
plt.show()
```

We can compare the rankings of the top 10 countries in terms of total GDP and GDP per capita to gain insights into the economic performance of these countries. It is possible that some countries may have high total GDPs due to their large populations, but their per capita GDP may be relatively low.

By comparing the rankings of these countries in terms of both total GDP and GDP per capita, we can gain a more complete understanding of their economic performance. For example, some countries may have high total GDPs but low per capita GDPs, indicating that their wealth is concentrated among a small group of individuals or industries. On the other hand, countries with high per capita GDPs may have lower total GDPs due to their smaller populations, but they may still be significant contributors to the global economy.

```python
Rank1 = df[['Country','Total_GDP ($)']].sort_values('Total_GDP ($)', ascending=False).reset_index()
Rank2 = df[['Country','GDP ($ per capita)']].sort_values('GDP ($ per capita)', ascending=False).reset_index()
Rank1 = pd.Series(Rank1.index.values+1, index=Rank1.Country)
Rank2 = pd.Series(Rank2.index.values+1, index=Rank2.Country)
Rank_change = (Rank2-Rank1).sort_values(ascending=False)
print('(Rank of Total GDP - Rank of GDP per capita)')
Rank_change.loc[top_gdp_countries.Country]
```

After comparing the rankings of the top 10 countries in terms of total GDP and GDP per capita, we can observe that the countries with high total GDPs may be different from those with high per capita GDPs.

For example, we can see that countries like China and India have much higher total GDPs than their per capita GDPs might suggest. This is likely due to their large populations, which allow them to generate significant economic output despite relatively low per capita incomes.

On the other hand, the United States is unique in that it ranks in the top 2 for both total and per capita GDPs. This indicates that the US has both a large overall economy and a relatively high level of economic prosperity for its citizens.
